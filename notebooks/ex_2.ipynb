```python
import pandas as pd
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.appName("Ex 2").getOrCreate()
```

```python
listings = spark.read.csv(
    "../data-src/listings.csv.gz",
    header=True,
    inferSchema=True,
    multiLine=True,
    sep=",",
    quote='"',
    escape='"',
    mode="PERMISSIVE",
)

listings.show(3)
```

```python
reviews = spark.read.csv(
    "../data-src/reviews.csv.gz",
    header=True,
    inferSchema=True,
    mode="PERMISSIVE",
    multiLine=True
)

reviews.show(3)
```

```python
# Q1: Number of reviews per listing
reviews.groupby("listing_id").count().alias("reviews per listing").show(truncate=False)
```

```python
spark.stop()
```
